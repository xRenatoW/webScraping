```python
import yfinance
```

```python
ticker = input("Digite o codigo da ação: ")
dados = yfinance.Ticker("PETR4.SA").history("6mo")
dados.Close
fechamento = dados.Close
fechamento.plot()
```

# Passo 2 - Criar análises

- Cotação Minima
- Cotação atual

```python
fechamento.min()
```

```python
maxima = fechamento.max()
minima = fechamento.min()
atual = fechamento[-1]
```

```python
import pyautogui
import pyperclip
```

```python
pyautogui.PAUSE = 2

pyautogui.hotkey("ctrl", "t")
pyperclip.copy("www.gmail.com")
pyautogui.hotkey("ctrl", "v")
pyautogui.hotkey("enter")
pyautogui.click(x=164, y=218)

pyperclip.copy("renato.x33@gmail.com")
pyautogui.hotkey("ctrl", "v")
pyautogui.hotkey("tab")
pyperclip.copy("Analise Diaria")
pyautogui.hotkey("ctrl", "v")
pyautogui.hotkey("tab")
mensagem = f""" 
Prezado Gestor,

Seguem as analises dos ultimos seis meses da ação {ticker}:

Cotação maxima R$ {round(maxima, 2)}
Cotação minima R$ {round(minima, 2)}
Cotação atual R$ {round(atual, 2)}

Qualquer duvida, estou a disposição!"""
pyperclip.copy(mensagem)
pyautogui.hotkey("ctrl", "v")
#pyperclip.copy("Analises diarias")
```
